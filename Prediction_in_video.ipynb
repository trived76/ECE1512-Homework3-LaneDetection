```python
from keras.models import load_model
model=load_model('lane_detector_model_UNet_3+2+2+1+1+1+1+1+1+1epochs_try_sigmoid.h5')
```

```python
def count_frame(vid_cap):
    
    shape1=0
    shape2=0
    success, image = vid_cap.read()
    if success==True:
        shape1=image.shape[0]
        shape2=image.shape[1]
    return (shape1,shape2)
```

```python
def making_video(input_filename,output_filename):
    
    cap=cv2.VideoCapture(input_filename)
    
    image_shape1,image_shape2=count_frame(cap)
    fourcc = cv2.VideoWriter_fourcc(*'MJPG')
    out = cv2.VideoWriter(output_filename, fourcc, 20.0, (image_shape2,image_shape1))
    
    while(cap.isOpened()):
        success, image = cap.read()
        if success==True:

            image1=cv2.resize(image,dsize=(256,256),interpolation=cv2.INTER_NEAREST)[:,:,::-1]
            image1=image1.astype(np.float32)/255.0
            temp=np.zeros((256,256)).astype(np.float32)/255.0

            # Setting white color as a highlighter of the detected lane
            temp[np.where((model.predict(image1[np.newaxis,:,:,:])[0].argmax(axis=2))==0)] = 1
            temp=cv2.resize(temp,dsize=(image_shape2,image_shape1),interpolation=cv2.INTER_NEAREST)
            image=image[:,:,::-1]
            image[np.where(temp==1)]=255
            out.write(image[:,:,::-1])
            print("It is running!")
            
        else:
            cap.release()
            out.release()
            cv2.destroyAllWindows()
            break
    return
```
